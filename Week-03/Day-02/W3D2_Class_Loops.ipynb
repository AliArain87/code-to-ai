## 2. `range()` — looping a fixed number of times

`range()` generates a sequence of numbers. It's perfect when you want to repeat something N times.

- `range(5)` → 0, 1, 2, 3, 4  *(starts at 0, stops **before** 5)*
- `range(1, 6)` → 1, 2, 3, 4, 5  *(start, stop)*
- `range(0, 10, 2)` → 0, 2, 4, 6, 8  *(start, stop, step)*

````python
list(range(1, 21))
````

````python
list(range(8))
````

````python
list(range(1,21,3))
````

````python
# Print numbers 0 to 4
for i in range(5):
    print(i)
````

````python
# Count from 1 to 5
for i in range(1, 6):
    print("Count:", i)
````

````python
# Step of 2 — even numbers
for i in range(0, 11, 2):
    print(i)
````

**Try it:** print the numbers from 10 down to 1 using `range(10, 0, -1)` (a negative step counts down).

````python
for i in range(10, 0, -1):
    print(i)
````

## 3. The `while` loop

A `for` loop runs a **known** number of times. A `while` loop runs **as long as a condition is True** — we may not know in advance how many times.

```
while condition:
    do this
```

Python checks the condition; if `True`, it runs the block, then checks again. It stops when the condition becomes `False`.

````python
count = 1

while count <= 5:
    print("Count is", count)
    count = count + 1   # VERY important: move toward stopping

print("Done!")
````

⚠️ **The most important line** is `count = count + 1`. It changes `count` so the condition eventually becomes `False`. Without it, the loop would run **forever** (an *infinite loop*).

## 4. Infinite loops — and how to avoid them

If the condition **never** becomes `False`, the loop never stops. This is a common beginner mistake.

```python
# DON'T RUN THIS — it never ends!
# count = 1
# while count <= 5:
#     print(count)      # count never changes → infinite loop
```

**Rule:** in every `while` loop, make sure something inside the loop moves the condition toward `False`.

*(If you ever get stuck in an infinite loop, press the ■ Stop button at the top, or Kernel → Interrupt.)*

````python
number = int(input("Enter number for table "))

while number < 11:
    print
````

## 5. `break` — stop the loop early

`break` immediately exits the loop, even if the condition is still `True`.

````python
for i in range(1, 100):
    print(i)
    if i == 20:
        print("Stopping early!")
        break
````

## 6. `continue` — skip to the next round

`continue` skips the rest of the current loop body and jumps to the next iteration.

````python
# Print numbers 1 to 10, but skip 5
for i in range(1, 20):
    if i == 5:
        continue    # skip printing 5
    print(i)
````

````python
for i in range(1,11):
    pass

print("hello")
````

## 7. Combining `if` inside loops

This is where yesterday and today come together. We can make decisions on each pass of a loop.

````python
numbers = [4, 7, 10, 3, 8, 15, 115, 112]

for n in numbers:
    if n % 2 == 0:
        print(n, "is even")
    else:
        print(n, "is odd")
````

````python
10 % 2
````

````python
# Sum only the even numbers
numbers = [4, 7, 10, 3, 8, 15]
total = 0

for n in numbers:
    if n % 2 == 0:
        total = total + n

print("Sum of even numbers:", total)
````

````python
# Sum only the even numbers
numbers = [4, 7, 10, 3, 8, 15]
total = 0

for n in numbers:
    if n % 2 != 0:
        total = total + n

print("Sum of even numbers:", total)
````

## 8. A common pattern — build up a result

Loops are often used to **accumulate** something: a total, a count, a new list.

````python

len('python')
````

````python
# Count how many words are longer than 4 letters
words = ["cat", "python", "AI", "banana", "sun"]
long_count = 0

for word in words:
    if len(word) > 4:
        long_count = long_count + 1

print("Words longer than 4 letters:", long_count)
````

## 9. Input validation pattern (`while` + `if`)

A very practical use of `while`: keep asking until the user gives valid input.

*(In Colab, `input()` opens a text box. Type a value and press Enter.)*

````python
# Keep asking until the user types a positive number
number = int(input("Enter a positive number: "))

while number <= 0:
    print("That's not positive. Try again.")
    number = int(input("Enter a positive number: "))

print("Thank you! You entered:", number)
````

This pattern — *ask, check, re-ask if wrong* — is exactly what we'll use in tomorrow's project. 👀
