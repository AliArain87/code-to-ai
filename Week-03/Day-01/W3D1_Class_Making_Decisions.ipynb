## 1. Quick recap — comparison operators

A comparison always gives back a **boolean**: `True` or `False`. Let's remind ourselves:

````python
age = 20

print(age > 18)      # is age greater than 18?
print(age == 20)     # is age exactly 20?
print(age != 15)     # is age NOT equal to 15?
print(age <= 18)     # is age less than or equal to 18?
````

These `True`/`False` values are the key. Today we use them to **choose** what our program does next.

## 2. The `if` statement

`if` runs a block of code **only when** its condition is `True`.

```
if condition:
    do this
```

Notice two things:
1. The line ends with a **colon** `:`
2. The next line is **indented** (4 spaces). This indentation tells Python *"this belongs inside the if"*.

````python
age = 20

if age >= 18:
    print("You are an adult.")
    print("You can vote.")

print("This line always runs (it is not indented).")
````

````python
age = 15

if age >= 18:
    print("You are an adult.")
    print("You can vote.")

print("This line always runs (it is not indented).")
````

**Try it:** change `age` to `15` and run again. The two indented lines are skipped, but the last line still runs.

Indentation is not optional in Python — it is how Python knows what is *inside* the `if`.

## 3. `if` / `else` — two paths

What if we want to do one thing when the condition is `True`, and **something else** when it is `False`? That's `else`.

````python
age = 15

if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")
````

Exactly **one** of the two blocks runs — never both, never neither. Change `age` and see.

## 4. `if` / `elif` / `else` — many paths

Sometimes there are more than two possibilities. `elif` (short for *else if*) lets us check more conditions in order.

Python checks each condition from top to bottom and runs the **first** one that is `True`. The rest are skipped.

````python
marks = 82

if marks >= 90: # 82 >= 90 False ---  ignore
    print("Grade: A+")
elif marks >= 80: # 82 >= 80 True
    print("Grade: A")
elif marks >= 70:
    print("Grade: B")
elif marks >= 60:
    print("Grade: C+")
elif marks >= 55:
    print("Grade: C")
else:
    print("Grade: F")
````

**Order matters!** Python stops at the first match. With `marks = 82`, it checks `>= 90` (False), then `>= 80` (True) → prints "Grade: A" and stops.

Try `marks = 95`, `marks = 55`, `marks = 70`.

## 5. Logical operators inside conditions

Remember `and`, `or`, `not`? They let us combine conditions.

- `and` → both sides must be True
- `or` → at least one side must be True
- `not` → flips True/False

````python
age = int(input("Enter your age: "))
has_ticket = True

print(f"User's age is {age}")

# BOTH must be true
if age >= 18 and has_ticket:
    print("Welcome to the show!")
else:
    print("Sorry, you cannot enter.")
````

````python
day = "Sunday"

# EITHER condition is enough
if day == "Saturday" or day == "Sunday":
    print("It's the weekend!")
elif day == "Monday":
    print("This is your first day on work")
else:
    print("It's a weekday.")
````

````python
name = "Ali"

print("hasnain" in name)
````

````python
is_raining = True

if not is_raining:
    print("Let's go outside!")
````

## 6. Membership in conditions — `in`

We can check if a value exists inside a collection (list, string, etc.) using `in`. This gives `True`/`False`, so it works perfectly in an `if`.

````python
fruits = ["Apple", "Banana", "Mango"]

# user_fruit = "orange"
user_fruit = input("What do you want sir? ")

if user_fruit in fruits:
    print(user_fruit, "is available!")
else:
    print("Sorry, we don't have that.")
````

````python
name = "Ali Hasnain"
name_2 = "Ali"
print(name == name_2)
````

````python
paragraph = "I am Ali Hasnain, and My friend name is Abdullah"
# print(type(name))

# 'in' also works on strings
if "Ali" in paragraph and "Abdullah" in paragraph:
    print("Found 'Ali' and 'Abdullah' in the name.")
````

## 7. Truthy & Falsy (intuition)

In an `if`, Python doesn't strictly need `True`/`False`. It treats some values as *"falsy"* (act like False) and everything else as *"truthy"* (act like True).

**Falsy values** (act like False):
- `0`
- `""` (empty string)
- `[]` (empty list)
- `None`

Everything else is **truthy**.

````python
name = ""  # empty string

if name:
    print("Hello,", name)
else:
    print("You didn't enter a name!")
````

````python
items = ["pen", "book"]

if items:
    print("Your cart has", len(items), "items.")
else:
    print("Your cart is empty.")
````

This is a clean way to check *"is there anything here?"* without writing `if len(items) > 0:`.
