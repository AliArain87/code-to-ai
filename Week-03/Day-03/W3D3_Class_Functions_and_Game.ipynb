## 4. `return` vs `print`

This is an important idea:

- `print()` just **shows** something on screen.
- `return` **gives a value back** to whoever called the function, so it can be stored and reused.

A returned value can go into a variable.

````python
def add(a, b):
    return a + b       # give the answer back
    # print(a+b)

````

````python
result = add(5,10)
print(result)
new_result = result * 2
print(new_result)
````

````python
result = add(3, 5)     # store the returned value
print("The result is:", result)

# We can even use it in more math
print("Double the result:", result * 2)
````

If we had used `print` inside `add` instead of `return`, we could **see** the answer but couldn't **reuse** it. `return` gives you the value to work with.

## 5. Local scope (light touch)

A variable created **inside** a function only exists inside that function. This is called its **scope**. Don't worry too much about this now — just know that variables inside a function stay "inside their own room".

````python
def calculate():
    secret = 42  
    double_secret = secret *2        # this 'secret' lives only inside calculate()
    return secret, double_secret

print(calculate())
# print(secret)   # <- this would cause an error: 'secret' is not defined out here
````

````python
# print(secret)
my_secret, double_secret = calculate()
````

````python
print(my_secret)
````

````python
print(my_secret)
````

````python
print(double_secret)
````

## 7. 🎮 PROJECT: Number-Guessing Game

**The idea:**
1. The computer picks a secret random number between 1 and 100.
2. The player keeps guessing.
3. After each guess, the program says **"Too high"** or **"Too low"**.
4. When the guess is correct, the game congratulates the player and tells them how many guesses they took.

This uses **everything** from this week: `random`, a `while` loop, `if`/`elif`/`else`, a counter, and functions. Let's build it step by step.

### Step 1 — Pick the secret number and set up

````python
import random
````

````python
secert_num = random.randint(1,50)
````

````python
print(secert_num)
````

````python
guess_user = int(input("Guess a number in between (1,50)"))
print(guess_user)
````

````python
if guess_user > secert_num:
    print("Guess lower")
elif guess_user < secert_num:
    print("Guess Higher")
else:
    print("Correct Guess")
````

````python
import random

secret_number = random.randint(1, 100)
guess_count = 0

print("🎮 Welcome to the Number-Guessing Game!")
print("I'm thinking of a number between 1 and 100.")
````

````python
print(secret_number)
````

### Step 2 — The guessing loop

We use `while True:` (an intentional loop that runs until we `break`). Each pass:
- ask for a guess
- count it
- compare it to the secret number

````python
while True:
    guess = int(input("Enter your guess: "))
    guess_count += 1

    if guess > secret_number:
        print("Guess lower")
    elif guess < secret_number:
        print("Guess Higher")
    else:
        print("Correct Guess", secret_number )
        print(f"Your used {guess_count} attempts") 
        break
````

````python
while True:
    guess = int(input("Enter your guess: "))
    guess_count = guess_count + 1

    if guess < secret_number:
        print("Too low! Try a higher number. ⬆️")
    elif guess > secret_number:
        print("Too high! Try a lower number. ⬇️")
    else:
        print("🎉 Correct! The number was", secret_number)
        print("You got it in", guess_count, "guesses!")
        break
````

That's a complete, working game! 🎉 Run Step 1 then Step 2 to play.

### Step 3 — Wrap it in a function (the clean version)

Now let's package the whole game into a function so it's reusable and tidy. This is the version you'll push to GitHub.

````python
import random

def play_game():
    secret_number = random.randint(1, 100)
    guess_count = 0

    print("🎮 Welcome to the Number-Guessing Game!")
    print("I'm thinking of a number between 1 and 100.")

    while True:
        guess = int(input("Enter your guess: "))
        guess_count = guess_count + 1

        if guess < secret_number:
            print("Too low! Try a higher number. ⬆️")
        elif guess > secret_number:
            print("Too high! Try a lower number. ⬇️")
        else:
            print("🎉 Correct! The number was", secret_number)
            print("You got it in", guess_count, "guesses!")
            break

````

````python
# Start the game
play_game()
````
