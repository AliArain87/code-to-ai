## 2. Reading — access by key
- `dict[key]` → value milti hai
- Agar key na ho to error. Safe tareeqa: `.get()`

```python
student = {"name": "Nabiha", "marks": 92}

print(student["name"])          # Nabiha
print(student.get("marks"))     # 92
print(student.get("grade"))     # None  -> error nahi, safe!
# print(student["grade"])
print(student.get("grade", "N/A"))  # default value de sakte hain
```

## 4. ⭐ Looping a dictionary
Teen tareeqe — keys, values, ya dono.

```python
marks = {"Math": 85, "English": 78, "Science": 92}

# Loop over keys (default)
for subjects in marks:
    print("Subject", subjects)
    print("Inside Loop")

print("Out of loop now")
```

```python
marks.keys()
```

```python
marks.values()
```

```python
marks.items()
```

```python
marks = {"Math": 85, "English": 78, "Science": 92}

# .items() -> key aur value dono ek saath (most useful!)
for subject, score in marks.items():
    print(subject, "->", score)
```

```python
marks = {"Math": 85, "English": 78, "Science": 92}

print(list(marks.keys()))     # ['Math', 'English', 'Science']
print(list(marks.values()))   # [85, 78, 92]
```

## 5. Sets — unique values, no duplicates 🎯
- Curly braces `{}` (jaise dict, lekin sirf values)
- Duplicates apne aap hat jaate hain
- Order guaranteed nahi hota

```python
s1 = {}
print(s1)
print(type(s1))
```

```python
s1 = set()
print(s1)
print(type(s1))
```

```python
S1 = {1,2,3,4,5}
print(S1)
print(type(S1))
```

```python
num = {1,1,1,2,4,4,5,4,6}
print(num)
```

```python
L1 = [1,2,4,4,5,5,6,7,6]
temp = set(L1)
print(temp)
print(type(temp))
L1 = list(temp)
print(L1)
print(type(L1))
```

```python
# Set se duplicates remove karna — super common trick!
numbers = [1, 2, 2, 3, 3, 3, 4]
unique = set(numbers)
print(unique)          # {1, 2, 3, 4}

names = ["Ali", "Sara", "Ali", "Bilal", "Sara"]
print(set(names))      # duplicates gone
```

```python
# Basic set operations
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a | b)   # union -> sab milake {1,2,3,4,5,6}
print(a & b)   # intersection -> common {3, 4}


print(a.union(b))
print(a.intersection(b))
```

```python
# Loop a set
for n in {10, 20, 30}:
    print(n)
```

## 7. 📒 Live build — Contact Book (in memory)
Aaj memory mein banayenge; **kal isay CSV file mein save karenge.**

```python
contacts = []

# Add a few contacts (har ek ek dictionary hai)
contacts.append({"name": "Ali",     "phone": "0342-3987710", "city": "Karachi"})
contacts.append({"name": "Noorulain","phone": "0301-5555555", "city": "Hyderabad"})
contacts.append({"name": "Fatimah","phone": "0301-5555555", "city": "Karachi"})

print("📒 My Contact Book")
print("=" * 35)

for c in contacts:
    print(c["name"], "-", c["phone"], "-", c["city"])

print("=" * 35)
print("Total contacts:", len(contacts))
print(type(contacts))
```

## ✅ Recap — aaj humne seekha
- **Dictionary** = `{key: value}` — data by *naam*, not position
- Access: `dict[key]` ya safe `dict.get(key)`
- Add/update: `dict[key] = value` · remove: `del dict[key]`
- Loop: `for k, v in dict.items():`
- **Set** = unique values, duplicates auto-remove: `set([...])`
- ⭐ **List of dictionaries** = rows of data → kal CSV banayenge
