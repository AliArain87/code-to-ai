## 1. The problem: too many variables 😵

Maan lo humein 5 doston ke naam store karne hain:

```python
# The painful way — one variable per friend
friend1 = "Farhan"
friend2 = "Nabiha"
friend3 = "Fatimah"
friend4 = "Noorulain"
friend5 = "Ali"

print(friend1, friend2, friend3, friend4, friend5)
# Problem: 100 doston ke liye 100 variables? Nahi! 😅
```

### The fix: a **list** — one box, many items 📦

```python
friends = ["Farhan", "Nabiha", "Fatimah", "Noorulain", "Ali"]
print(friends)
print(type(friends))   # <class 'list'>
```

## 2. Reading a list — indexing
- Counting starts at **0** (not 1!)
- Negative index counts from the end (`-1` = last item)

```python
friends = ["Farhan", "Nabiha", "Fatimah", "Noorulain", "Ali"]

print(friends[0])    # first  -> Farhan
print(friends[1])    # second -> Nabiha
print(friends[-1])   # last   -> Ali
print(friends[-2])   # second last -> Noorulain
```

### Slicing — grabbing a *range* `list[start:stop]` (stop not included)

```python
numbers = [10, 20, 30, 40, 50, 60]

print(numbers[0:3])   # [10, 20, 30]  -> index 0,1,2
print(numbers[2:])    # [30, 40, 50, 60]
print(numbers[:2])    # [10, 20]
print(numbers[-2:])   # [50, 60]
```

## 3. Changing a list ✏️
Lists are **mutable** — we can add, change, and remove items.

```python
cart = ["milk", "eggs"]

cart.append("bread")        # add to the end
print(cart)                 # ['milk', 'eggs', 'bread']

cart.insert(1, "butter")    # add at position 1
print(cart)                 # ['milk', 'butter', 'eggs', 'bread']

cart[0] = "almond milk"     # change an item
print(cart)
```

```python
cart = ["milk", "butter", "eggs", "bread"]

cart.remove("butter")   # remove by value
print(cart)             # ['milk', 'eggs', 'bread']

last = cart.pop()       # remove + return the last item
print(last)             # bread
print(cart)             # ['milk', 'eggs']
```

## 4. Useful list tools 🛠️
`len()` · `in` · `sort()` · `sorted()`

```python
fruits = ["banana", "apple", "cherry"]

print(len(fruits))            # 3  -> how many items
print("apple" in fruits)      # True
print("mango" in fruits)      # False

fruits.sort()                 # sorts the list itself (A-Z)
print(fruits)                 # ['apple', 'banana', 'cherry']

nums = [4, 1, 3, 2]
print(sorted(nums))           # [1, 2, 3, 4]  -> returns a NEW sorted list
print(nums)                   # [4, 1, 3, 2]  -> original unchanged
```

## 5. ⭐ NEW: the `for item in collection:` loop
Yeh hai is week ki sabse badi cheez! Ek list ke har item par baari baari kaam karna.

> `for` ek-ek item uthata hai aur indented line(s) chalata hai — har item ke liye.

```python
friends = ["Farhan", "Nabiha", "Fatimah", "Noorulain", "Ali"]

for name in friends:
    print("Hello,", name)

# 'name' har baar agli value ban jaata hai. Magic! ✨
```

```python
# Loop ke andar hisaab bhi kar sakte hain
prices = [100, 250, 80, 420]

for p in prices:
    print("Price with tax:", p * 1.1)
```

> 💡 **Indentation matters.** Loop ke andar wali line 4 spaces andar hoti hai. Wahi line har item ke liye chalti hai.

## 6. Tuples — lists that **can't** change 🔒
- Round brackets `()` instead of square `[]`
- Immutable: ek baar bana, badal nahi sakte
- Use when data should stay fixed (e.g., coordinates, RGB colour, days of week)

```python
point = (3, 5)          # x, y
print(point[0])         # 3
print(point[1])         # 5

rgb = (255, 0, 0)       # red
print("Red value:", rgb[0])

# Tuples ko loop bhi kar sakte hain
for value in rgb:
    print(value)
```

```python
# Tuple change karne ki koshish -> ERROR (yeh expected hai!)
point = (3, 5)
# point[0] = 10   # uncomment karo to dekho: TypeError
print("Tuples stay safe and unchanged 🔒")
```

## 7. 🎬 Live build — Favourite Movies printer
Sab kuch jo aaj seekha, ek choti si cheez mein.

```python
movies = ["Inception", "Interstellar", "The Martian", "Dune"]

print("🎬 My Favourite Movies")
print("=" * 25)

for movie in movies:
    print("-", movie)

print("=" * 25)
print("Total movies:", len(movies))
```
