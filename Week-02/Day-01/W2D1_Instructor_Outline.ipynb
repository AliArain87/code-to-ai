## 3. Changing a list ✏️
Lists are **mutable** — we can add, change, and remove items.

```python
cart = ["milk", "eggs"]
print(cart)

cart.append("bread")        # add to the end
print(cart)  


cart.append(3)
print(cart)               # ['milk', 'eggs', 'bread']

cart.insert(1, "butter")    # add at position 1
print(cart)                 # ['milk', 'butter', 'eggs', 'bread']

# cart[0] = "almond milk"     # change an item
# print(cart)
```

```python
cart[2] = 'Almond Milk'
cart
```

```python
cart
```

```python
cart.insert(1, 'banana')
print(cart)
```

```python
cart.insert(3, 'shake')
print(cart)
```

```python
others = ['apple', 2]
others
```

```python
cart
```

```python
cart.extend(others)
cart
```

```python
cart = ["milk", "butter", "eggs", "bread"]

cart.remove("butter")   # remove by value
print(cart)             # ['milk', 'eggs', 'bread']

last = cart.pop()       # remove + return the last item
print(last)             # bread
print(cart)             # ['milk', 'eggs']
```

## 5. ⭐ NEW: the `for item in collection:` loop
Yeh hai is week ki sabse badi cheez! Ek list ke har item par baari baari kaam karna.

> `for` ek-ek item uthata hai aur indented line(s) chalata hai — har item ke liye.

```python
friends = ["Farhan", "Nabiha", "Fatimah", "Noorulain", "Ali"]

for name in friends:
    print("Hello,", name)

# 'name' har baar agli value ban jaata hai. Magic! ✨
```

```python
# Loop ke andar hisaab bhi kar sakte hain
prices = [100, 250, 80, 420]

for p in prices:
    print("Price with tax:", p * 1.1)
```

> 💡 **Indentation matters.** Loop ke andar wali line 4 spaces andar hoti hai. Wahi line har item ke liye chalti hai.

## 6. Tuples — lists that **can't** change 🔒
- Round brackets `()` instead of square `[]`
- Immutable: ek baar bana, badal nahi sakte
- Use when data should stay fixed (e.g., coordinates, RGB colour, days of week)

```python
l2=  [1,2]
print(l2)
print(type(l2))
```

```python
t1 = ()
print(t1)
print(type(t1))
```

```python
points = (2,4,5)
print(points)
print(type(points))
```

```python
l1 = [1, 2, 4, 5]
l1[1] = 'Apple'
print(l1)
```

```python
points
```

```python
points[1] = 'Apple'
points
```

```python
points[3]
```

```python
list1 = [1]
print(list1)
print(type(list1))
```

```python
t1 = (1,)
print(t1)
print(type(t1))
```

```python
t1 = (1,2)
print(t1)
print(type(t1))
```

```python
t2 = (1,2,4,5,6)
print(t2)
```

```python
t2[-2]
```

```python
point = (3, 5)          # x, y
print(point[0])         # 3
print(point[1])         # 5

rgb = (255, 0, 0)       # red
print("Red value:", rgb[0])

# Tuples ko loop bhi kar sakte hain
for value in rgb:
    print(value)
```

```python
# Tuple change karne ki koshish -> ERROR (yeh expected hai!)
point = (3, 5)
# point[0] = 10   # uncomment karo to dekho: TypeError
print("Tuples stay safe and unchanged 🔒")
```

## 7. 🎬 Live build — Favourite Movies printer
Sab kuch jo aaj seekha, ek choti si cheez mein.

```python
movies = ["Inception", "Interstellar", "The Martian", "Dune"]

print("🎬 My Favourite Movies")
print("=" * 25)

for movie in movies:
    print("-", movie)

print("=" * 25)
print("Total movies:", len(movies))
```
