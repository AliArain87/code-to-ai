## 1. Aggregations (0–12 min)

An **aggregation** takes many numbers and returns one summary number.
NumPy gives us these directly:
- `.sum()` — total
- `.mean()` — average
- `np.median()` — middle value
- `.min()` / `.max()` — smallest / largest

````python
import numpy as np

scores = np.array([70, 85, 90, 60, 75, 95, 80])
````

````python
print(scores)
````

````python
print("sum   :", scores.sum())
````

````python
print("mean  :", scores.mean())
````

````python
print("median:", np.median(scores))
````

````python
print("min   :", scores.min())
````

````python
print("max   :", scores.max())
````

## 2. Spread: std & var (12–25 min)

The **mean** tells us the center. But two groups can have the same mean and still look very different — one tightly packed, one spread out.

- **Standard deviation** (`.std()`) measures how spread out the numbers are.
- **Variance** (`.var()`) is the standard deviation squared.

Small std = numbers are close together. Large std = numbers are spread out.

````python
tight = np.array([49, 50, 51, 50, 50])   # all near 50
spread = np.array([10, 90, 30, 70, 50])  # far apart

print("tight  mean:", tight.mean(),  "std:", round(tight.std(), 2))
print("spread mean:", spread.mean(), "std:", round(spread.std(), 2))
# Same-ish mean, very different spread!
````

````python
import numpy as np

a = np.array([48, 49, 50, 51, 52])

print("Mean:", np.mean(a))
print("Standard Deviation:", np.std(a))
````

````python
import numpy as np

b = np.array([10, 30, 50, 70, 90])

print("Mean:", np.mean(b))
print("Standard Deviation:", np.std(b))
````

## 3. The `axis` idea (25–40 min) — taught visually

For a 2D array (a table), we can aggregate in three ways:

```
             col0  col1  col2
   row0  ->   [ 1,   2,   3 ]
   row1  ->   [ 4,   5,   6 ]
```

- `axis=0` → go **down** the columns (one result per column)
- `axis=1` → go **across** the rows (one result per row)
- no axis → one number for the whole array

Tip: `axis=0` collapses rows, `axis=1` collapses columns.

````python
grid = np.array([[1, 2, 3],
                 [4, 5, 6]])

print("whole sum :", grid.sum())          # 21
````

````python
print("col sums  :", grid.sum(axis=0))    # [5 7 9]  (down columns)
````

````python


print("row sums  :", grid.sum(axis=1))    # [6 15]   (across rows)
````

````python
# Same idea with mean — useful for real tables
print("column averages:", grid.mean(axis=0))
print("row averages   :", grid.mean(axis=1))
````

## 4. Boolean masks & filtering (40–55 min)

Remember comparison operators from Week 3? On arrays, a comparison returns an array of **True/False** — a **mask**.

We can then use that mask to **pull out** only the elements we want. This is one of the most useful tools in data analysis.

````python
temps = np.array([31, 28, 35, 33, 26, 37, 30])

mask = temps > 32          # a True/False array
print("mask      :", mask)
````

````python
print("hot days  :", temps[mask])       # only values where mask is True
````

````python
# We can count and combine conditions too
print("how many hot days:", (temps > 32).sum())   # True counts as 1
````

````python
print("mild days        :", temps[temps < 30])
````

## 5. `np.where` for labeling (55–66 min)

`np.where(condition, value_if_true, value_if_false)` builds a new array by choosing a value for each element based on a condition. Great for **labeling** data.

````python
temps = np.array([31, 28, 35, 33, 26, 37, 30])

labels = np.where(temps > 32, "HOT", "OK")
print(labels)

````

````python
# It also works with numbers (e.g., cap values)
capped = np.where(temps > 34, 34, temps)   # anything above 34 becomes 34
print(capped)
````

## 6. Reshaping (66–75 min)

Sometimes we need to change an array's shape without changing its data.
- `reshape(rows, cols)` → rearrange into a new shape (total size must match)
- `flatten()` → squash a 2D array back into 1D

````python
a = np.arange(1, 13)     # 12 numbers: 1..12
print("original:", a)

grid = a.reshape(3, 4)   # 3 rows, 4 cols  (3*4 = 12)
print(grid)

grid_2 = a.reshape(4,3)
print(grid_2)
````
