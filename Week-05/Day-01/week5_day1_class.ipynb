## 1. Why Pandas?

In Week 4 we used **NumPy**. NumPy arrays are fast, but they have one big limit:
**everything is a number, and columns have no names.**

Real data is not like that. Real data looks like a spreadsheet:

| name | age | city | fare |
|------|-----|------|------|
| Ali | 25 | Karachi | 7.25 |
| Sara | 31 | Lahore | 71.28 |

Different types in different columns. Every column has a **name**.

**Pandas** gives us exactly this: a table in Python.

- NumPy = a calculator for numbers
- Pandas = a spreadsheet you can program

Pandas is the tool data scientists spend most of their day inside. From today until the end of this course, it is our main tool.

## 2. The Series — a labeled 1D column

A **Series** is one column of data.
It has two parts:
- the **values** (the data itself)
- the **index** (the label of each value)

Think of it as a list that remembers names.

````python
list1 = [1,2,3,4,5]
print(list1)
````

````python
# Creating a Series from a list
marks = pd.Series([85, 92, 78, 95])
print(marks)
````

Look at the output carefully.

The left column `0 1 2 3` is the **index** (automatic, like list positions).
The right column is the **values**.
`dtype: int64` tells us the data type.

Now let's give the index our own labels.

````python
marks = pd.Series([85, 92, 78, 95], index=["Ali", "Sara", "Bilal", "Ayesha"])
print(marks)
````

````python
# The three things every Series has
print("Values:", marks.values)
print("Index :", marks.index)
print("Dtype :", marks.dtype)
````

### Creating a Series from a dictionary

A dictionary already has keys and values — which is exactly index and values.
Pandas understands this automatically.

````python
marks = pd.Series({"Ali": 85, "Sara": 92, "Bilal": 78, "Ayesha": 95})
print(marks)
````

### Indexing a Series: by label and by position

````python
# By label
print("Sara's marks:", marks["Sara"])

````

````python
# By position (like a list)
print("First student's marks:", marks.iloc[0])
print("First student's marks:", marks.iloc[1])
print("First student's marks:", marks.iloc[0:2])
````

````python
# Multiple labels at once
marks_2 = marks[["Ali", "Bilal"]]
print(marks_2)
print(marks[["Ali", "Ayesha"]])
````

````python
marks
````

````python
# Math works on the whole Series at once, like NumPy
print(marks + 5)
````

## 3. The DataFrame — rows and columns

A **DataFrame** is a full table.

The easiest mental model:
> A DataFrame is a dictionary of Series. Each column is a Series. They all share the same index.

Let's build one from a dictionary of lists.

````python
students = pd.DataFrame({
    "name":  ["Ali", "Sara", "Bilal", "Ayesha", "Hassan", "Asfand"],
    "age":   [25, 31, 22, 28, 35, 22],
    "city":  ["Karachi", "Lahore", "Karachi", "Islamabad", "Lahore", "Umerkot"],
    "marks": [85, 92, 78, 95, 66, 95]
})

print(students)
````

````python
students
````

Notice we wrote just `students` at the end of the cell instead of `print(students)`.
In a notebook, this shows a nicely formatted table. Use it for DataFrames.

### The shape of a DataFrame

````python
print("Shape (rows, columns):", students.shape)
print("Columns:", students.columns)
print("Index  :", students.index)
print()
print("Data types of each column:")
print(students.dtypes)
````

## 5. The 5-command ritual

Every time you meet a new dataset, run these five commands. Every time. No exceptions.

1. `.head()` — what does it look like?
2. `.shape` — how big is it?
3. `.info()` — what are the columns and types?
4. `.describe()` — what do the numbers look like?
5. `.isna().sum()` — what is missing?

This takes 30 seconds and saves you hours.

````python
# 1. head() — the first 5 rows
titanic.head()
````

````python
# You can ask for any number of rows
titanic.head(3)
````

````python
# tail() — the last rows
titanic.tail(3)
````

````python
# sample() — random rows. Useful because head() can be misleading if data is sorted.
titanic.sample(10)
````

````python
# 2. shape
print("Rows, Columns:", titanic.shape)
print("Number of passengers:", titanic.shape[0])
````

````python
# 3. info()
titanic.info()
````

### Understanding the Titanic columns

| Column | Meaning |
|--------|---------|
| `survived` | 0 = died, 1 = survived |
| `pclass` | Ticket class as a number: 1, 2, 3 |
| `sex` | male / female |
| `age` | Age in years |
| `sibsp` | Number of siblings / spouses aboard |
| `parch` | Number of parents / children aboard |
| `fare` | Ticket price |
| `embarked` | Port they boarded from (C, Q, S) |
| `class` | Ticket class as text: First, Second, Third |
| `who` | man / woman / child |
| `alone` | True if travelling alone |

Some columns repeat the same information in different forms (`pclass` and `class`). That is normal in real data.

````python
# 4. describe() — statistics for every numeric column at once
titanic.describe()
````

Read this table row by row:
- `count` — how many values are present
- `mean` — the average
- `std` — standard deviation (spread), from Week 4
- `min` / `max` — the extremes
- `25%` / `50%` / `75%` — the quartiles. `50%` is the median.

Already we can see something: `age` has a count of 714 while other columns have 891.
**177 ages are missing.**

````python
# 5. isna().sum() — count the missing values in each column
titanic.isna().sum()
````

This is our first look at missing data. We are only **counting** it today.
**Fixing** missing data is the whole topic of Week 6.

## 6. Selecting columns

Two forms, and the difference matters:

- `df['age']` — one set of square brackets → returns a **Series**
- `df[['age', 'fare']]` — two sets → returns a **DataFrame**

The inner brackets are just a **list of column names**. That is why there are two.

````python
# Single column -> Series
ages = titanic["age"]
print(type(ages))
print(ages.head(10))
````

````python
# Multiple columns -> DataFrame
titanic[["age", "fare", "survived"]].head(10)
````

````python
# Even one column in a list gives a DataFrame
subset = titanic[["age"]]
print(type(subset))
subset.head()
````

## 7. Quick summaries

Once you have a column (a Series), you can summarize it instantly.

````python
print("Average age  :", titanic["age"].mean())
print("Oldest       :", titanic["age"].max())
print("Youngest     :", titanic["age"].min())
print("Median age   :", titanic["age"].median())
print("Total fare   :", titanic["fare"].sum())
````

````python
# The mean of a 0/1 column is the percentage of 1s. A very useful trick.
print("Survival rate:", titanic["survived"].mean())
````

That number means about 38% of passengers survived. Because `survived` is 0 or 1,
the average **is** the survival rate. Remember this trick — we use it a lot on Day 3.

````python
# value_counts() — how many of each category?
titanic["sex"].value_counts()
````

````python
titanic
````

````python
titanic['class'].value_counts()
````

````python
# As percentages instead of counts
titanic["class"].value_counts(normalize=True)
````

````python
# unique() and nunique()
print("Unique ports:", titanic["embarked"].unique())
print("How many unique ports:", titanic["embarked"].nunique())
print("Unique classes:", titanic["class"].unique())
print("Unique classes:", titanic["class"].nunique())
````
