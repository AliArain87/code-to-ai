## 1 · `input()` — talking to the user (15 min)

`input()` pauses the program and lets the user type something.

**THE golden rule:** `input()` **always returns a `str`** — even if they type a number. To do math, **cast** it with `int()` or `float()`.


```python
name = input("What is your name?")
print(f"Nice to meet you, {name}! 👋👱‍♂️")
```

```python
num1 = int(input("Please enter num1 "))
num2 = int(input("Please enter num2 "))

print(f"You entered num1 {num1}")
print(f"You entered num2 {num2}")

print(f"{num1 + num2}")

print(f"The type of {num1} and {num2}  is {type(num1)} and {type(num2)}")
```

```python
num1 = float(input("Please enter num1 "))
num2 = float(input("Please enter num2 "))

print(f"You entered num1 {num1}")
print(f"You entered num2 {num2}")

print(f"{num1 + num2}")

print(f"The type of {num1} and {num2}  is {type(num1)} and {type(num2)}")
```

```python
a = input("First number: ")   # user types 5  -> "5" (string!)
b = input("Second number: ")  # user types 3  -> "3"
print(a + b)   # ❗ prints "53", not 8 — strings get joined
```

```python
# ✅ The fix: cast to numbers
a = int(input("First number: "))
b = int(input("Second number: "))
print(f"Sum = {a + b}")
```

## 2 · Types of Python Operators (25 min)

Operators are symbols that *do something* to values. We'll cover **5 families**.


### 2.1 Arithmetic Operators ➕ (the math ones)
| Op | Means | Example | Result |
|----|-------|---------|--------|
| `+` | add | `7 + 2` | `9` |
| `-` | subtract | `7 - 2` | `5` |
| `*` | multiply | `7 * 2` | `14` |
| `/` | divide | `7 / 2` | `3.5` |
| `//` | floor divide | `7 // 2` | `3` |
| `%` | modulus (remainder) | `7 % 2` | `1` |
| `**` | power | `7 ** 2` | `49` |


```python
a = int(input("a "))
b = int(input("b "))
print(f"{a} + {b} = {a + b}")
print(f"{a} / {b} = {a / b}")
print(f"{a} // {b} = {a // b}   (floor divide)")
print(f"{a} % {b} = {a % b}     (remainder)")
print(f"{a} ** {b} = {a ** b}   (power)")
```

### 2.2 Assignment Operators 📥 (store & update)
`=` assigns. Shortcuts update a variable using its own value.
| Op | Same as | 
|----|---------|
| `x += 3` | `x = x + 3` |
| `x -= 3` | `x = x - 3` |
| `x *= 3` | `x = x * 3` |
| `x /= 3` | `x = x / 3` |


```python
score = 10
print(f"Start: {score}")
score += 5      # add 5 score = score + 5
print(f"After += 5: {score}")
score *= 2      # double it
print(f"After *= 2: {score}")
```

```python
X = 5
print(X)
# X = X + 2
X += 2
print(X)
```

### 2.3 Comparison Operators ⚖️ (ask a yes/no question → `True`/`False`)
We'll just **print the answers** today.
| Op | Means |
|----|-------|
| `==` | equal to |
| `!=` | not equal |
| `>` `<` | greater / less |
| `>=` `<=` | greater-equal / less-equal |


```python
a = 3
b = 3

print(f"{14 == 14}")
```

```python
a = "ali"
b = "ali"

print(f"{a == b}")
```

```python
# !=
x = 21
b = 21

print(x != b)
```

```python
a = "ali"
b = "ali"

print(f"{a != b}")
```

```python
print(f"{"ali" != "ali"}")
```

```python
print(f"3 > 3     -> {3 > 3}")
print(f"9 >= 9    -> {9 >= 9}")
```

9 >= 3 # True
3 >= 3 # False

```python
# print({10>=9})
print(f"{3 >= 3}")
```

```python
print(f"{9 < 3}")
```

```python
print(f"{9 <= 3}")
```

```python
print(f"{9 <= 9}")
```

```python
print(f"10 == 10  -> {10 == 10}")
print(f"10 != 5   -> {10 != 5}")
print(f"7 > 3     -> {7 > 3}")
print(f"4 >= 9    -> {4 >= 9}")
```

### 2.4 Logical Operators 🔗 (`and`, `or`, `not`)
Combine `True`/`False` values. Again, just **print the results** for now.
| Op | True when… |
|----|------------|
| `and` | **both** sides True |
| `or` | **at least one** side True |
| `not` | flips True ↔ False |


```python
print(5 > 3)
```

```python
print((10 > 2))
```

```python
print(1 > 3)
```

```python
print(f"(5 > 3) and (10 > 2) -> {(1 > 3) and (1 > 2)}")
```

```python
print(f"(5 > 3) or  (1 > 9)  -> {(5 >= 3) or (1 >= 9)}")
```

```python
print(f"not (5 > 3)          -> {not (5 > 3)}")
```

```python
print(f"{not (1 > 4)}")
```

### 2.5 Membership Operators 🔎 (`in`, `not in`)
Check if something is **inside** a string (or collection — more in Week 2).


```python
name = "Ali"
print(f"Is 'A' in name? -> {'A' in name}")
print(f"Is 'z' in name? -> {'z' in name}")
print(f"'x' not in name -> {'x' not in name}")
```

## 3 · BUILD: Interactive Calculator (20 min) 🧮

The Week 1 project. We take two numbers and print **all** results at once (straight-line — no menu/`if` yet).


```python
print("🧮 Welcome to my Calculator!")
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

print("\n--- Results ---")
print(f"{num1} + {num2} = {num1 + num2}")
print(f"{num1} - {num2} = {num1 - num2}")
print(f"{num1} * {num2} = {num1 * num2}")
print(f"{num1} / {num2} = {num1 / num2:.2f}")
print(f"{num1} ** {num2} = {num1 ** num2}")
```

```python
x = int(input("Enter value for x "))
y = int(input("Enter value for y "))

# addition
add = x + y

# sub
sub = x - y

# divide
div = x / y

# Power
power = x ** y

# Floor
flr = x // y 

# Quotition
rem = x % y


print(f"The result of {x} + {y} = {add}")
print(f"The result of {x} - {y} = {sub}")
print(f"The result of {x} / {y} = {div:.3f}")
print(f"The result of {x} ** {y} = {power}")
print(f"The result of {x} // {y} = {flr}")
print(f"The result of {x} % {y} = {rem}")
```
